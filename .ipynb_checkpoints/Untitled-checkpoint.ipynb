```python
from matplotlib import pyplot as plt
# 设置图片大小，以及单位像素点
plt.figure(figsize=(20,8),dpi=80)
x = range(2,26,2)
y = [15, 12, 13, 14, 31, 32, 28, 13, 16, 19, 12, 14]
plt.plot(x, y)
# 设置X轴坐标
plt.xticks(x)
#plt.savefig(""./sig_size.png")
plt.show()
```

```python
import random
import matplotlib
plt.rcParams['font.family'] = 'SimHei'
plt.figure(figsize=(20,8),dpi=80)
x = range(0,120)
y = [random.randint(20,35) for i in range(120)]
x_label=['10点{}分'.format(i) for i in range(60)]
x_label+=['11点{}分'.format(i) for i in range(60)]
# 旋转角度
plt.xticks(x[::10],x_label[::10],rotation=45)
plt.title("时间天气")
plt.xlabel("时间")
plt.ylabel("气温")
plt.plot(x, y)
# 设置X轴坐标
#plt.xticks(range(0,120,5))
#plt.savefig(""./sig_size.png")
plt.show()
```

```python
plt.rcParams['font.family'] = 'SimHei'
plt.figure(figsize=(20,8),dpi=80)
year = range(11,31)
size1 = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1]
size2 = [0,1,0,8,3,3,3,2,2,3,3,6,1,3,2,3,1,1,2,3]
_year_label = ['{}岁'.format(i) for i in year]
plt.xticks(year[::2],_year_label[::2],rotation=45)
plt.yticks(range(0,10))
plt.plot(year,size1,label="自己",color="red",linestyle=":",linewidth=5)
plt.plot(year,size2,label="同桌",color="cyan",linestyle="--",linewidth=5)
plt.title("交男女朋友走势")
# plt.xlabel("年龄")
# plt.ylabel("个数")
#plt.ylim(min(size),max(size))
# 添加右上角的图例,loc标识放在哪个位置
plt.legend(loc = 0)
plt.grid(alpha=1,linestyle=":")
plt.show()
```

```python
import random
import matplotlib
y_3=[random.randint(20,30) for i in range(31)]
y_10=[random.randint(20,30) for i in range(31)]
plt.figure(figsize=(20,8),dpi=80)
x_3= range(1,32)
x_10=range(51,82)
# 绘制散点图
plt.scatter(x_3,y_3,label="3月份")
plt.scatter(x_10,y_10,label="10月份")
_x=list(x_3)+list(x_10)
_x_label=["3月{}号".format(i) for i in x_3]
_x_label+=["10月{}号".format(i-50) for i in x_10]
plt.xticks(_x[::3],_x_label[::3],rotation=45)
plt.legend(loc=1)

plt.show()
```
