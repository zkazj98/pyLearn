```python
import yfinance as yf
import datetime
import numpy as np
import matplotlib.pyplot as plt

msft = yf.Ticker("MSFT")
hist = msft.history(start='2024-04-01',end='2024-04-23')
hist['diff'] = hist['Close'].diff()
hist['Signal']=np.where(hist['diff']>0,1,0)
hist

plt.figure(figsize=(10,5))
hist['Close'].plot(linewidth=2,color='k',grid=True)
plt.scatter(hist['Close'].loc[hist.Signal==1].index,hist['Close'][hist.Signal==1],
           marker='v',s=80,c='g')
plt.scatter(hist['Close'].loc[hist.Signal==0].index,hist['Close'][hist.Signal==0],
           marker='^',s=80,c='r')
plt.show()
```

```python
import yfinance as yf
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

zgpa = yf.Ticker("AAPL")

hist = zgpa.history(start='2024-04-01',end='2024-04-23')
initcash=20000
zgpa_signal=pd.DataFrame(index=hist.index)
zgpa_signal['price']=hist['Close']
zgpa_signal['diff']=zgpa_signal['price'].diff()
zgpa_signal['signal']=np.where(zgpa_signal['diff'] >=0 ,0,1)
zgpa_signal['order']=zgpa_signal['signal'].diff()*100
zgpa_signal['stock']=zgpa_signal['order']*zgpa_signal['price']
zgpa_signal['cash']=initcash-(zgpa_signal['order'].diff()*zgpa_signal['price']).cumsum()
zgpa_signal['total']=zgpa_signal['stock']+zgpa_signal['cash']
plt.figure(figsize=(10,6))
plt.plot(zgpa_signal['total'])
plt.plot(zgpa_signal['order'].cumsum()*zgpa_signal['price'],'--',label='stock value')
plt.grid()
plt.legend(loc='center right')
zgpa_signal
```
